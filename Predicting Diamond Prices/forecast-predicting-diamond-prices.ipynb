### Context
This classic dataset contains the prices and other attributes of almost 54,000 diamonds. It's a great dataset for beginners learning to work with data analysis and visualization.

### Content
- **price**: Price in US dollars (\$326--\$18,823)
- **carat**: Weight of the diamond (0.2--5.01)
- **cut**: Quality of the cut (Fair, Good, Very Good, Premium, Ideal)
- **color**: Diamond colour, from J (worst) to D (best)
- **clarity**: A measurement of how clear the diamond is (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))
- **x**: Length in mm (0--10.74)
- **y**: Width in mm (0--58.9)
- **z**: Depth in mm (0--31.8)
- **depth**: Total depth percentage = z / mean(x, y) = 2 * z / (x + y) (43--79)
- **table**: Width of top of diamond relative to widest point (43--95)

```python
# Function to Plot Hist and Box
def plot_histograms_and_boxplots(data, columns, color='#FF7F50'):
    num_plots = len(columns)
    fig, axes = plt.subplots(num_plots, 2, figsize=(15, num_plots*5))
    
    for i, col in enumerate(columns):
        sns.histplot(data[col], ax=axes[i, 0], color=color)
        axes[i, 0].set_title(f'Histogram of {col}')
        
        sns.boxplot(data[col], ax=axes[i, 1], color=color)
        axes[i, 1].set_title(f'Boxplot of {col}')
        
    plt.tight_layout()
    plt.show()
    
# Plot 
print_boxed_zigzag_heading('Numerical features')
plot_histograms_and_boxplots(tr_d,['carat','depth', 'table', 'price', 'x', 'y',
       'z'])
```

```python
# Remove Outliers
def remove_outliers_iqr(df):
    # Selecting only float features
    float_columns = df.select_dtypes(include=['float64']).columns

    # Calculate first quartile (Q1)
    q1 = df[float_columns].quantile(0.25)
    # Calculate third quartile (Q3)
    q3 = df[float_columns].quantile(0.75)
    # Calculate IQR (Interquartile Range)
    iqr = q3 - q1

    # Identify outliers
    outlier_mask = ((df[float_columns] < (q1 - 1.5 * iqr)) | (df[float_columns] > (q3 + 1.5 * iqr)))
    # Count outliers in each column
    num_outliers = outlier_mask.sum()

    # Print the number of outliers
    print_boxed_zigzag_heading("Number of outliers detected:\n")
    print(num_outliers)
    
    # Remove outliers
    df_cleaned = df[~outlier_mask.any(axis=1)]

    # Count outliers after removal
    num_outliers_cleaned = ((df_cleaned[float_columns] < (q1 - 1.5 * iqr)) | (df_cleaned[float_columns] > (q3 + 1.5 * iqr))).sum()
    print_boxed_zigzag_heading("Number of outliers after removal:\n")
    print(num_outliers_cleaned)

    return df_cleaned

# Exceuting Function
tr_d = remove_outliers_iqr(tr_d)
```

```python
# Defining the categorical columns to encode
CAT_COL_E = ['cut','color','clarity']

# Function to encode columns using Label Encoding
def LabelEncode(data, columns):
    # Initialize LabelEncoder
    L_E = LabelEncoder()
    
    # Encode categorical columns
    for col in columns:
        data[col] = L_E.fit_transform(data[col])
    
    # Return the encoded data
    return data

# Encoder Train  
tr_d = LabelEncode(tr_d, CAT_COL_E)
print_boxed_zigzag_heading('Data is Encoded Successfully')
```

# <p style="text-align:center;"> Model || Building  </p>

```python
# # # =================================================================================================================
# # #                         X < y 
# # #================================================================================================================== 
X_T = tr_d.drop('carat', axis=1)
y_T = tr_d['carat']

# # # =================================================================================================================
# # #                         Train < Test Split
# # #================================================================================================================== 
X_TR, X_TE, Y_TR, Y_TE = train_test_split(X_T, y_T, test_size=0.1, random_state=42)

# # # =================================================================================================================
# # #                         Scale Data 
# # #================================================================================================================== 
scaler = MinMaxScaler()
#USing Loop
for col in X_TR.columns:
    X_TR[col] = scaler.fit_transform(X_TR[[col]])
for col in X_TE.columns:
    X_TE[col] = scaler.fit_transform(X_TE[[col]])

# # # =================================================================================================================
# # #                         Shapes < 
# # #================================================================================================================== 
print_boxed_zigzag_heading(f"Training set shape - X: {X_TR.shape}, y: {Y_TR.shape}")
print_boxed_zigzag_heading(f"Testing set shape - X: {X_TE.shape}, y: {Y_TE.shape}")
```

```python
#Sequentail Model
model = Sequential()

#Imput Layer 
model.add(Dense(10,activation='relu',input_dim=9, kernel_regularizer=l1(0.01)))
#Hidden Layers 
model.add(Dense(10,activation='relu'))
#OutputLayer
model.add(Dense(1,activation='linear'))

#Compile
model.compile(loss='huber',optimizer='Adam',metrics='mse')

# Summary 
model.summary()

#Fit 
history = model.fit(X_TR,
                    Y_TR,
                    epochs=15,
                    validation_split=0.2)
```
